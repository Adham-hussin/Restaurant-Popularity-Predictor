# requirement 1 :

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df = df[0]
```

getting the data from wekipedia

```python
df.drop(df[df.Borough == "Not assigned"].index, inplace=True)
```

droping the unassigned-Borough rows

```python
for n in range(103):
    if df.iloc[n,2] == "Not assigned" :
        df.iloc[n, 2] = df.loc[n, 1]
```

assigning unassigned neighborhood values to the corresponding borough

# requirement 3 :

```python
df4 = df3[df3["Borough"].str.contains("Toronto", regex= False)]
```

```python
df4
```

```python
torontoAnalysis = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,Neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(Neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat,lng],radius=5,popup=label,color='pink',
    fill=True,fill_color='red',fill_opacity=0.6).add_to(torontoAnalysis)
torontoAnalysis
```

```python
cluster = KMeans(n_clusters= 5,init="k-means++",n_init=10,random_state=0)
```

```python
df5 = df4.iloc[: , 3:5]
```

```python
cluster.fit(df5)
```

```python
cluster.labels_
```

```python
df4.insert(0, 'cluster labels', cluster.labels_)
```

```python
df4
```

```python
torontoAnalysis2= folium.Map(location=[43.651070,-79.347015],zoom_start=12)
c = ['Blue','Red','Green','cyan']
```

```python
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['cluster labels']):
    label = folium.Popup(' Cluster ' + str(cluster))
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=c[cluster-1],
        fill=True,
        fill_color=c[cluster -1],
        fill_opacity=0.6).add_to(torontoAnalysis2)
       
torontoAnalysis2
```
