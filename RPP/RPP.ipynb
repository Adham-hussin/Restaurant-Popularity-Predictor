```python
import pandas as pd
import numpy as np
import requests
import json
import geopy
from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values
from sklearn.cluster import KMeans
import folium
import matplotlib.pyplot as plt
import seaborn as sns
```

importing the libraries

```python
CLIENT_ID = 'removed' # your Foursquare ID
CLIENT_SECRET = 'removed' # your Foursquare Secret
VERSION = '20180604'
LIMIT = 30
search_query = 'restaurant'
radius = 1000
```

api information

```python
address = '102 North End Ave, New York, NY'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

getting the location of north end eve

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
url
```

building the url to call the api

```python
res = requests.get(url).json()
```

```python
with open('data2.json') as json_data:
    newyork_data = json.load(json_data)
```

```python
df = pd.DataFrame(newyork_data)
```

```python
df2 = df['response']['venues']
```

```python
df3 = pd.json_normalize(df2)
```

requesting, saving, calling and normalizing the data

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
df3
```

```python
df4 = df3[["id","location.lat","location.lng","delivery.id", "name"]]
df4.head()
```

cleaning the data 

```python
venues_ids= df3['id']
ratings=[]
for venue_id in venues_ids.values.tolist():
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID, CLIENT_SECRET, VERSION)
    result = requests.get(url).json()
    with open('data3.json', 'a+') as file:
        json.dump(result, file)
    try:
        venues_rating=result['response']['venue']['rating']
        ratings=ratings+[venues_rating]
    except:
        print('This venue has not been rated yet.')
        ratings = ratings + ["NaN"]
```

requesting every rating of found restaurants 

```python
df4.insert(1,"ratings",ratings)
```

```python
df4['delivery.id'] = df4['delivery.id'].fillna(0)
```

```python
for i in range(30) :
    if df4.iloc[i, 4] == 0 :
        df4.iloc[i, 4] = 0
    else:
        df4.iloc[i, 4] = 1
```

```python
for i in range(30) :
    if df4.iloc[i, 1] == "NaN" :
        df4.iloc[i, 1] = 0
```

visualising the data on a map

```python
df10 = df4[["ratings", "delivery.id"]]
```

```python
df10.rename(columns={"delivery.id": "delivery"})
df10.head()
```

taking the data that will be used for clustering

```python
cluster = KMeans(n_clusters= 4,init="k-means++",n_init=10,random_state=0)
cluster.fit(df10)
```

fitting the model

```python
df4.insert(0, 'cluster labels', cluster.labels_)
```
